```python
from datasets import load_dataset
import pandas as pd

# ❶  Load just a few rows so it’s quick
ds = load_dataset("MedRAG/textbooks")
```

```python
print("First 10 instances from the dataset:")
for i, example in enumerate(ds['train'].select(range(10))):
    print(f"\nInstance {i+1}:")
    for key, value in example.items():
        # Truncate long text fields for better readability
        if isinstance(value, str) and len(value) > 100:
            value = value[:100] + "..."
        print(f"{key}: {value}")
```
